### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
file_path = "mnt/data/crypto_market_data.csv"
df = pd.read_csv("crypto_market_data.csv")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Create a DataFrame with the scaled data
df_scaled = df.copy()

# Copy the crypto names from the original DataFrame
# Set the coin_id column as index
if 'coin_id' in df.columns:
 df_scaled.set_index('coin_id', inplace=True)

# Display the scaled DataFrame
print(df_scaled.head())
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,12))
print(k_values)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

if df_scaled.select_dtypes(include='object').shape[1] > 0:
    df_scaled = pd.get_dummies(df_scaled)


# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
k_range = range(1, 12)
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)

print(inertia_values)
```

```python
pip install matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary with the data to plot the Elbow curve
elbow_data ={
    'Number of Clusters': range(1, 12),
    'Inertia': inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Plot the Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(elbow_df['Number of Clusters'], elbow_df['Inertia'], marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(8, 6))
plt.plot(range(1, 12), inertia_values, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

plt.title('Elbow Method: Inertia vs. Number of Clusters', fontsize=14)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia', fontsize=12)

plt.grid(True)

plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:**  From the plot, I believe the best value for K is 3. The sharp drop occurs between k=1 and k=3 and after k=3. The decrease in inertia becomes more gradual.

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 3
kmeans = KMeans(n_clusters=best_k, random_state=42)
kmeans.fit(df_scaled)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
clusters = kmeans.predict(df_scaled)

# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_copy = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_copy['Cluster'] = clusters

# Display the copy of the scaled DataFrame
print(df_scaled_copy)
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = df_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",
    hover_cols=["crypto_name"]
)

print(scatter_plot)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_transformed = pca.fit_transform(df_scaled_copy.drop(columns=['Cluster']))
pca_df = pd.DataFrame(pca_transformed, columns=["PC1", "PC2", "PC3"], index=df_scaled_copy.index)
pca_df['Cluster'] = clusters

# View the scaled PCA data
print(pca_df)
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
# YOUR CODE HERE!
inertia_values = []
scaled_inertia_values = []
k_values = list(range(1, 11))

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled_copy.drop(columns=['Cluster']))
    inertia_values.append(kmeans.inertia_)
    
    scaler = StandardScaler()
    df_scaled_optimized = scaler.fit_transform(df_scaled_copy.drop(columns=['Cluster']))
    kmeans_opt = KMeans(n_clusters=k, random_state=42)
    kmeans_opt.fit(df_scaled_optimized)
    scaled_inertia_values.append(kmeans_opt.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(k_values, inertia_values, marker='o', label='Without Optimization')
plt.plot(k_values, scaled_inertia_values, marker='s', linestyle='dashed', label='With Optimization')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve Analysis: With and Without Optimization')
plt.legend()
plt.show()
```
